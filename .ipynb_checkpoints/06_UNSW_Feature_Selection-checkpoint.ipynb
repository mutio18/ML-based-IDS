```python
print("="*60)
print(" LOADING UNSW-NB15 DATA FOR FEATURE ANALYSIS")
print("="*60)

# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_selection import mutual_info_classif
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print(" Libraries imported")

# Load the parquet files
train_raw = pd.read_parquet('data/UNSW/UNSW_NB15_training-set.parquet')
test_raw = pd.read_parquet('data/UNSW/UNSW_NB15_testing-set.parquet')

print(f" Raw training data shape: {train_raw.shape}")
print(f" Raw test data shape: {test_raw.shape}")
print(f"\nColumns in raw data: {list(train_raw.columns)[:10]}...")

# Load preprocessed data (without UNSW_ prefix)
X_train = pd.read_csv('data/processed/X_train_final.csv')  # Note: X_train_final.csv
y_train = pd.read_csv('data/processed/y_train.csv').values.ravel()  # Note: y_train.csv

print(f"\n Preprocessed training data shape: {X_train.shape}")
print(f" Preprocessed labels shape: {y_train.shape}")
print(f"\nFeature names: {list(X_train.columns)[:10]}...")
```

```python
print("="*60)
print(" FEATURE CORRELATION WITH ATTACK LABEL")
print("="*60)

# Calculate correlation of each feature with target
correlations = []
for col in X_train.columns:
    corr = abs(X_train[col].corr(pd.Series(y_train)))
    correlations.append({'feature': col, 'correlation': corr})

# Sort by correlation
corr_df = pd.DataFrame(correlations).sort_values('correlation', ascending=False)

print("Top 15 features most correlated with attacks:")
print(corr_df.head(15).to_string(index=False))

# Visualize
plt.figure(figsize=(10, 8))
top_corr = corr_df.head(15)
plt.barh(top_corr['feature'], top_corr['correlation'], color='steelblue')
plt.xlabel('Correlation with Attack Label')
plt.title('Top 15 Features by Correlation')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig('results/UNSW/feature_correlation.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
print("="*60)
print(" MUTUAL INFORMATION ANALYSIS")
print("="*60)

# Calculate mutual information (can capture non-linear relationships)
mi_scores = mutual_info_classif(X_train, y_train, random_state=42)
mi_df = pd.DataFrame({
    'feature': X_train.columns,
    'mutual_info': mi_scores
}).sort_values('mutual_info', ascending=False)

print("Top 15 features by mutual information:")
print(mi_df.head(15).to_string(index=False))

# Visualize
plt.figure(figsize=(10, 8))
top_mi = mi_df.head(15)
plt.barh(top_mi['feature'], top_mi['mutual_info'], color='purple')
plt.xlabel('Mutual Information Score')
plt.title('Top 15 Features by Mutual Information')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig('results/UNSW/feature_mutual_info.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
print("="*60)
print(" RANDOM FOREST FEATURE IMPORTANCE")
print("="*60)

# Train quick Random Forest on sample (faster)
X_sample, _, y_sample, _ = train_test_split(
    X_train, y_train, train_size=0.3, random_state=42, stratify=y_train
)

rf = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
rf.fit(X_sample, y_sample)

# Get feature importances
rf_importance = pd.DataFrame({
    'feature': X_train.columns,
    'importance': rf.feature_importances_
}).sort_values('importance', ascending=False)

print("Top 15 features by Random Forest:")
print(rf_importance.head(15).to_string(index=False))

# Visualize
plt.figure(figsize=(10, 8))
top_rf = rf_importance.head(15)
plt.barh(top_rf['feature'], top_rf['importance'], color='green')
plt.xlabel('Importance Score')
plt.title('Top 15 Features by Random Forest')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig('results/UNSW/feature_rf_importance.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
print("="*60)
print(" ATTACK DISTRIBUTION BY PROTOCOL")
print("="*60)

# Use raw data for protocol analysis
protocol_attack = pd.crosstab(train_raw['proto'], 
                              train_raw['label'] > 0, 
                              margins=True, 
                              margins_name="Total")

protocol_attack.columns = ['Normal', 'Attack', 'Total']
print("Attack counts by protocol:")
print(protocol_attack)

# Calculate attack percentages
protocol_pct = pd.crosstab(train_raw['proto'], 
                           train_raw['label'] > 0, 
                           normalize='index') * 100
protocol_pct.columns = ['Normal %', 'Attack %']
print("\nAttack percentage by protocol:")
print(protocol_pct.sort_values('Attack %', ascending=False))

# Visualize
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Count plot
protocol_counts = protocol_attack.drop('Total').sort_values('Attack', ascending=False).head(10)
protocol_counts[['Normal', 'Attack']].plot(kind='bar', ax=axes[0], 
                                          color=['green', 'red'])
axes[0].set_title('Top 10 Protocols - Normal vs Attack Counts')
axes[0].set_xlabel('Protocol')
axes[0].set_ylabel('Count')
axes[0].tick_params(axis='x', rotation=45)

# Percentage plot
protocol_pct_sorted = protocol_pct.sort_values('Attack %', ascending=False).head(10)
protocol_pct_sorted['Attack %'].plot(kind='bar', ax=axes[1], color='red')
axes[1].axhline(y=50, color='black', linestyle='--', alpha=0.5, label='50% line')
axes[1].set_title('Top 10 Protocols - Attack Percentage')
axes[1].set_xlabel('Protocol')
axes[1].set_ylabel('Attack %')
axes[1].tick_params(axis='x', rotation=45)
axes[1].legend()

plt.tight_layout()
plt.savefig('results/UNSW/protocol_attack_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
print("="*60)
print(" ATTACK DISTRIBUTION BY SERVICE")
print("="*60)

# Service analysis
service_attack = pd.crosstab(train_raw['service'], 
                             train_raw['label'] > 0, 
                             margins=True, 
                             margins_name="Total")
service_attack.columns = ['Normal', 'Attack', 'Total']

# Calculate percentages
service_pct = pd.crosstab(train_raw['service'], 
                          train_raw['label'] > 0, 
                          normalize='index') * 100
service_pct.columns = ['Normal %', 'Attack %']

print("Top 15 most attacked services:")
service_stats = pd.DataFrame({
    'Total': service_attack['Total'],
    'Attack_Count': service_attack['Attack'],
    'Attack_%': service_pct['Attack %']
}).drop('Total').sort_values('Attack_Count', ascending=False).head(15)

print(service_stats.to_string())

# Visualize
plt.figure(figsize=(12, 8))
services = service_stats.index
attack_pcts = service_stats['Attack_%'].values
colors = ['red' if p > 70 else 'orange' if p > 40 else 'green' for p in attack_pcts]

plt.barh(services, attack_pcts, color=colors)
plt.xlabel('Attack Percentage (%)')
plt.title('Top 15 Services by Attack Count - Attack %')
plt.axvline(x=50, color='black', linestyle='--', alpha=0.5, label='50%')
plt.legend()
plt.tight_layout()
plt.savefig('results/UNSW/service_attack_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
print("="*60)
print(" ATTACK CATEGORIES DISTRIBUTION")
print("="*60)

attack_cats = train_raw['attack_cat'].value_counts()
attack_cats_pct = train_raw['attack_cat'].value_counts(normalize=True) * 100

attack_df = pd.DataFrame({
    'Count': attack_cats,
    'Percentage': attack_cats_pct
})

print("Attack categories:")
print(attack_df)

# Visualize
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Bar plot
attack_cats.plot(kind='bar', ax=axes[0], color='coral')
axes[0].set_title('Attack Categories - Count')
axes[0].set_xlabel('Attack Type')
axes[0].set_ylabel('Count')
axes[0].tick_params(axis='x', rotation=45)

# Pie chart
axes[1].pie(attack_cats.values, labels=attack_cats.index, autopct='%1.1f%%')
axes[1].set_title('Attack Categories - Percentage')

plt.tight_layout()
plt.savefig('results/UNSW/attack_categories.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
print("="*60)
print(" FEATURE IMPORTANCE COMPARISON")
print("="*60)

# Combine top features from different methods
top_features = pd.DataFrame({
    'Correlation': corr_df.head(10)['feature'].values,
    'Mutual Info': mi_df.head(10)['feature'].values,
    'Random Forest': rf_importance.head(10)['feature'].values
})

print("Top 10 features from each method:")
print(top_features.to_string(index=False))

# Find consensus features
from collections import Counter

all_top_features = (list(corr_df.head(20)['feature']) + 
                   list(mi_df.head(20)['feature']) + 
                   list(rf_importance.head(20)['feature']))

consensus = Counter(all_top_features).most_common(10)
print("\n Consensus Top 10 Features (appear in multiple methods):")
for i, (feature, count) in enumerate(consensus, 1):
    print(f"{i:2d}. {feature} (appears in {count} methods)")
```

```python
print("="*60)
print(" FEATURE ANALYSIS SUMMARY")
print("="*60)

print("Key Findings:")
print("-" * 40)
print("1. Top correlated features with attacks:")
for i, row in corr_df.head(5).iterrows():
    print(f"   • {row['feature']}: {row['correlation']:.4f}")

print("\n2. Most suspicious protocols (high attack %):")
suspicious_protos = protocol_pct.sort_values('Attack %', ascending=False).head(5)
for proto, row in suspicious_protos.iterrows():
    if proto != 'Total':
        print(f"   • {proto}: {row['Attack %']:.1f}% attacks")

print("\n3. Most attacked services:")
top_services = service_stats.head(5)
for service, row in top_services.iterrows():
    print(f"   • {service}: {row['Attack_Count']:.0f} attacks ({row['Attack_%']:.1f}%)")

print("\n4. Most common attack categories:")
for cat, count in attack_cats.head(3).items():
    print(f"   • {cat}: {count} ({count/len(train_raw)*100:.1f}%)")

print("\n Feature analysis complete!")
print(" All visualizations saved to results/UNSW/")
```
